## Question 2

As discussed in the lecture, a significant challenge, particularly when dealing with data that have many columns, is ensuring each column gets encoded correctly. 

This is particularly true with data columns that are ordered categoricals (ordinals) vs unordered categoricals. Unordered categoricals should be one-hot encoded, however this can significantly increase the number of features and creates features that are highly correlated with each other.

Determine how many total features would be present, relative to what currently exists, if all string (object) features are one-hot encoded. Recall that the total number of one-hot encoded columns is `n-1`, where `n` is the number of categories.

```python
# Select the object (string) columns
obj_columns = [col for col in df.columns if df.dtypes[col] == object]    #modify this line
obj_columns

#mask = df.dtypes == np.object
#obj_columns = df.columns[mask]
#obj_columns
```

```python
df[obj_columns]
```

```python
# Determine how many extra columns would be created
extra_col = (df[obj_columns]
            .apply(lambda x: x.nunique())
            .sort_values(ascending=False))
print("Extra Columns:\n",extra_col)
# No need to encode if there is only one value as that will be converted to an integer using label encoding
updated_extra_col = extra_col.loc[extra_col > 1]   
# Number of one-hot columns is one less than the number of categories as index starts from 0
updated_extra_col = updated_extra_col - 1
print("Updated Extra Columns:\n",updated_extra_col) 
# This is 215 columns, assuming the original ones are dropped. 
# This is quite a few extra columns!

#print the total number of extra column that could be created
print(updated_extra_col.sum())    #modify this line
```

## Question 3

Let's create a new data set where all of the above categorical features will be one-hot encoded. We can fit this data and see how it affects the results.

* Used the dataframe `.copy()` method to create a completely separate copy of the dataframe for one-hot encoding
* On this new dataframe, one-hot encode each of the appropriate columns and add it back to the dataframe. Be sure to drop the original column.
* For the data that are not one-hot encoded, drop the columns that are string categoricals.

For the first step, numerically encoding the string categoricals, either Scikit-learn;s `LabelEncoder` or `DictVectorizer` can be used. However, the former is probably easier since it doesn't require specifying a numerical value for each category, and we are going to one-hot encode all of the numerical values anyway. (Can you think of a time when `DictVectorizer` might be preferred?)

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Copy of the data
df_ohc = df.copy()
x=0
#print(df_ohc)
# The encoders
le = LabelEncoder()
ohc = OneHotEncoder()

for col in updated_extra_col.index:
    
    # Integer encode the string categories
    data_int_encode = le.fit_transform(df_ohc[col]).astype(np.int)     #converting each string col to int                         #modify this line
    # Remove the original column from the dataframe
    df_ohc = df_ohc.drop(col, axis = 1)   #modify this line
    
    # One hot encode the data--this returns a sparse array
    new_df = ohc.fit_transform(data_int_encode.reshape(-1,1))    #modify this line
    
    # Create unique column names
    new_cols = new_df.shape[1]
    #print(new_cols)

    column_names = ['_'.join([col, str(x)]) for x in range(new_cols)]
    print(column_names)
    
     # Create the new dataframe
    ohc_new_df = pd.DataFrame(new_df.toarray(), columns = column_names, index = df_ohc.index)
    
    # Append the new data to the dataframe
    df_ohc = pd.concat([df_ohc,ohc_new_df ], axis = 1)   #modify this line
    print(df_ohc.head(1))
    
```

## Question 4

* Create train and test splits of both data sets. To ensure the data gets split the same way, use the same `random_state` in each of the two splits.
* For each data set, fit a basic linear regression model on the training data. 
* Calculate the mean squared error on both the train and test sets for the respective models. Which model produces smaller error on the test data and why?

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

error_dataframe = list()

X_df  = df.drop('SalePrice',axis=1)   #modify this line
y_df = df['SalePrice']   #modify this line

X_ohc  = df_ohc.drop('SalePrice',axis=1)   #modify this line
y_ohc = df_ohc['SalePrice']    #modify this line

#splitting data to train and test
X_train, X_test, y_train, y_test = train_test_split(X_df,y_df , test_size = 0.3 , random_state = 42)  #modify this line
X_ohc_train, X_ohc_test, y_ohc_train, y_ohc_test = train_test_split(X_ohc,y_ohc , test_size =0.3 , random_state = 42)  #modify this line

#fitting regular data
LR = LinearRegression()
LR.fit(X_train, y_train)     #modify this line
y_pred_train = LR.predict(X_train)   #modify this line
y_pred_test = LR.predict(X_test)    #modify this line

#####################################
# Fitting encoded data
LR = LinearRegression()
LR.fit(X_ohc_train,y_ohc_train)        #modify this line
y_ohc_pred_train = LR.predict(X_ohc_train)    #modify this line
y_ohc_pred_test = LR.predict(X_ohc_test)     #modify this line



 #modify the line below
error_dataframe.append(pd.Series({'Train': mean_squared_error(y_train,y_pred_train),
                               'Test' : mean_squared_error(y_test,y_pred_test)},
                              name = 'No encoding'))


 #modify the line below
error_dataframe.append(pd.Series({'Train': mean_squared_error(y_ohc_train,y_ohc_pred_train),
                               'Test' : mean_squared_error(y_ohc_test,y_ohc_pred_test)},
                              name = 'One-hot encoding'))

error_dataframe = pd.concat(error_dataframe, axis=1)
error_dataframe
```

The model that we genrated without one-hot encoding gives better result because it does not have overfitting issue. The error value of one-hot encoding on test data are very high because it is overfitted.

## Question 5

For each of the data sets (one-hot encoded and not encoded):

* Scale the all the non-hot encoded values using one of the following: `StandardScaler`, `MinMaxScaler`, `MaxAbsScaler`.
* Compare the error calculated on the test sets

Be sure to calculate the skew (to decide if a transformation should be done) and fit the scaler on *ONLY* the training data, but then apply it to both the train and test data identically.

```python
# Mute the setting wtih a copy warnings
pd.options.mode.chained_assignment = None
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler
# defining a dictionary
scalers = {'standardscal': StandardScaler(),
           'minmaxscal' : MinMaxScaler(),
           'maxabcscal' : MaxAbsScaler()   
          }
training_test_sets = {
    'not encoded' : (X_train, y_train, X_test, y_test),
    'one-hot encoded' :  (X_ohc_train, y_ohc_train, X_ohc_test, y_ohc_test)
}
mask_for_float = X_train.dtypes == np.float64
float_col = X_train.columns[mask_for_float]

#model initialization
LR = LinearRegression()

#iterate over all the possible combinations and gets the err
err = {}
for enc_label, (_X_train, _y_train, _X_test, _y_test) in training_test_sets.items():
    for scaler_label, scaler in scalers.items():
        trainingset = _X_train.copy()
        testset = _X_test.copy()
        trainingset[float_col] = scaler.fit_transform (trainingset[float_col])   #modify this line
        testset[float_col] = scaler.transform (testset[float_col])   #modify this line
        LR.fit(trainingset,_y_train)  #modify this line
        predictions = LR.predict(testset)
        key = enc_label + ' - ' + scaler_label + 'scaling'
        err[key] = mean_squared_error(_y_test, predictions)
        
err = pd.Series(err)
```
